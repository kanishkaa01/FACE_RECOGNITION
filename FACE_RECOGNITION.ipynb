Image Processing

```python
setA = os.listdir("/content/model/E")
setB = os.listdir("/content/model/B")
setC = os.listdir("/content/model/D")
```

```python
limit = 100
A_images = [None]*limit
j=0
for i in setA:
    if (j<limit):
        A_images[j] = imread("/content/model/E/"+i)
        j+=1
    else:
        break
```

```python
limit = 100
B_images = [None]*limit
j=0
for i in setB:
    if (j<limit):
        B_images[j] = imread("/content/model/B/"+i)
        j+=1
    else:
        break
```

```python
limit = 100
C_images = [None]*limit
j=0
for i in setC:
    if (j<limit):
        C_images[j] = imread("/content/model/D/"+i)
        j+=1
    else:
        break
```

```python
C_gray[50].shape
```

```python
len_of_images_A=len(A_gray)
```

```python
len_of_images_B=len(B_gray)
```

```python
len_of_images_C=len(C_gray)
```

```python
image_size_A=A_gray[22].shape
image_size_A
```

```python
image_size_B=B_gray[20].shape
image_size_B
```

```python
image_size_C=C_gray[50].shape
image_size_C
```

Flatten

```python
flatten_size_A=image_size_A[0]*image_size_A[1]
flatten_size_A
```

```python
flatten_size_B=image_size_B[0]*image_size_B[1]
flatten_size_B
```

```python
flatten_size_C=image_size_C[0]*image_size_C[1]
flatten_size_C
```

```python
for i in range(len_of_images_A):
    A_gray[i] = np.ndarray.flatten(A_gray[i]).reshape(flatten_size_A,1)
```

```python
A_gray = np.dstack(A_gray)
A_gray = np.rollaxis(A_gray, axis=2, start=0)
A_gray.shape
```

```python
A_gray = A_gray.reshape(len_of_images_A, flatten_size_A)
A_gray.shape
```

```python
for i in range(len_of_images_B):
    B_gray[i] = np.ndarray.flatten(B_gray[i]).reshape(flatten_size_B,1)
```

```python
B_gray = np.dstack(B_gray)
B_gray = np.rollaxis(B_gray, axis=2, start=0)
B_gray.shape
```

```python
B_gray = B_gray.reshape(len_of_images_B, flatten_size_B)
B_gray.shape
```

```python
for i in range(len_of_images_C):
    C_gray[i] = np.ndarray.flatten(C_gray[i]).reshape(flatten_size_C,1)
```

```python
C_gray = np.dstack(C_gray)
C_gray = np.rollaxis(C_gray, axis=2, start=0)
C_gray.shape
```

```python
C_gray = C_gray.reshape(len_of_images_C, flatten_size_C)
C_gray.shape
```

```python
from sklearn.utils import shuffle
```

```python
images_indexed = shuffle(combine_2).reset_index()
```

```python
images_indexed
```

```python
drop_images = images_indexed.drop(['index'],axis =1)
```

```python
drop_images
```

Test And Train 

```python
x = drop_images.values[:,:-1]
```

```python
y = drop_images.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

Accuracy

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
